# Justificativa para Altera√ß√µes nos Tipos de Dados

## Motivo das Altera√ß√µes

- **Descri√ß√£o**: As altera√ß√µes nos tipos de dados foram realizadas para garantir a consist√™ncia e a precis√£o dos dados de acordo com o metadado fornecido e as necessidades anal√≠ticas do projeto.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo do seaborn (para consist√™ncia com gr√°ficos anteriores)
sns.set_style("whitegrid")
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['axes.labelsize'] = 10



# Passo 3: Criar vari√°veis derivadas relevantes para as perguntas de neg√≥cio
# 3.1. Crescimento absoluto de pl√°stico (para pergunta 1)
df['Plastic_Growth_tons'] = df['Plastic_Waste_2060_tons'] - df['Plastic_Waste_2015_tons']

# 3.2. Taxa de aumento de pl√°stico (para an√°lise de tend√™ncias)
df['Plastic_Growth_Rate'] = (df['Plastic_Growth_tons'] / df['Plastic_Waste_2015_tons']) * 100

# 3.3. Risco relativo baseado em Policy_Strength (para pergunta 5)
df['Relative_Risk'] = df['Risk_Score_Change'] / (df['Policy_Strength_2060'] + 0.01)  # +0.01 para evitar divis√£o por zero

# Passo 4: Verificar outliers nas vari√°veis quantitativas (m√©todo IQR)
print("\nVerifica√ß√£o de Outliers (M√©todo IQR)")
quantitative_vars = [
    'Plastic_Waste_2015_tons', 'Plastic_Waste_2060_tons', 'Plastic_Growth_tons',
    'Population_2015', 'Population_2060',
    'GDP_per_capita_2015', 'GDP_per_capita_2060',
    'Plastic_to_River_2060_tons',
    'Policy_Strength_2015', 'Policy_Strength_2060',
    'Waste_Collection_Rate_2015', 'Waste_Collection_Rate_2060',
    'Risk_Score_Change', 'Plastic_Growth_Rate', 'Relative_Risk'
]

for col in quantitative_vars:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    print(f"\nColuna: {col}")
    print(f"Limite Inferior: {lower_bound:.2f}, Limite Superior: {upper_bound:.2f}")
    print(f"N√∫mero de Outliers: {len(outliers)}")
    if len(outliers) > 0:
        print(f"Valores dos Outliers (at√© 10): {outliers.tolist()[:10]}")

# Passo 5: Padronizar e criar vari√°veis categ√≥ricas
# 5.1. Criar Plastic_Risk_Category (baixo, m√©dio, alto) baseado em Plastic_to_River_2060_tons
print("\nCria√ß√£o de Plastic_Risk_Category")
df['Plastic_Risk_Category'] = pd.cut(df['Plastic_to_River_2060_tons'],
                                   bins=[0, df['Plastic_to_River_2060_tons'].quantile(0.33),
                                         df['Plastic_to_River_2060_tons'].quantile(0.66),
                                         df['Plastic_to_River_2060_tons'].max()],
                                   labels=['Baixo', 'M√©dio', 'Alto'],
                                   include_lowest=True)
print("Valores √∫nicos de Plastic_Risk_Category:", df['Plastic_Risk_Category'].unique())

# 5.2. Criar High_Plastic_Growth (bin√°rio: 1 para alta taxa de crescimento, 0 para baixa)
print("\nCria√ß√£o de High_Plastic_Growth")
# Considerar alta taxa de crescimento como > 50% (baseado na amostra)
df['High_Plastic_Growth'] = (df['Plastic_Growth_Rate'] > 50).astype(int)
print("Valores √∫nicos de High_Plastic_Growth:", df['High_Plastic_Growth'].unique())

# 5.3. Criar Policy_Strength_Level
print("\nCria√ß√£o de Policy_Strength_Level")
df['Policy_Strength_Level'] = pd.cut(df['Policy_Strength_2060'],
                                   bins=[0, 0.3, 0.7, 1.0],
                                   labels=['Baixo (<0.3)', 'M√©dio (0.3-0.7)', 'Alto (>0.7)'])
print("Valores √∫nicos de Policy_Strength_Level:", df['Policy_Strength_Level'].unique())

# Passo 6: Verificar valores √∫nicos das vari√°veis qualitativas
print("\nValores √∫nicos das vari√°veis qualitativas:")
qualitative_vars = ['Country', 'Continent', 'River_Name']
for var in qualitative_vars:
    print(f"\n{var}:")
    print(df[var].unique())

# Passo 7: Exibir resumo do DataFrame atualizado
print("\nResumo do DataFrame ap√≥s Corre√ß√µes")
print(df.info())
print("\nPrimeiras 5 Linhas do DataFrame Atualizado")
print(df.head())
```

## Descri√ß√£o do M√©todo IQR (Intervalo Interquartil)

### Defini√ß√£o
- O **Intervalo Interquartil (IQR)** √© uma medida estat√≠stica utilizada para identificar valores at√≠picos (outliers) em um conjunto de dados. Ele √© calculado como a diferen√ßa entre o terceiro quartil (Q3, 75¬∫ percentil) e o primeiro quartil (Q1, 25¬∫ percentil), ou seja, \( IQR = Q3 - Q1 \).
- Os limites para detectar outliers s√£o definidos como:
  - **Limite Inferior**: \( Q1 - 1.5 \times IQR \)
  - **Limite Superior**: \( Q3 + 1.5 \times IQR \)
- Valores fora desses limites s√£o considerados outliers, indicando dados que se desviam significativamente da distribui√ß√£o central.

### Aplica√ß√£o no Projeto
- O m√©todo foi aplicado √†s vari√°veis quantitativas relevantes do dataset "River_Plastic_Waste_Risk_Scenarios_2015_2060.csv", incluindo tanto as originais (ex.: `Plastic_Waste_2015_tons`, `Population_2015`) quanto as derivadas (ex.: `Plastic_Growth_tons`, `Relative_Risk`).
- Para cada vari√°vel, foram calculados os quartis (Q1 e Q3), o IQR, e os limites inferior e superior. Em seguida, identificou-se o n√∫mero de outliers e os valores espec√≠ficos (at√© 10 exemplos).

### Resultados da Aplica√ß√£o
- Vari√°veis como `Plastic_Waste_2015_tons` (73 outliers), `Plastic_Waste_2060_tons` (64 outliers), e `Plastic_to_River_2060_tons` (147 outliers) apresentaram valores at√≠picos significativos.
- Outras vari√°veis, como `Population_2015` e `GDP_per_capita_2015`, n√£o apresentaram outliers, indicando uma distribui√ß√£o mais uniforme.

## Justificativa para a Utiliza√ß√£o do M√©todo IQR

### Necessidade
- **Detec√ß√£o de Dados At√≠picos**: A an√°lise preditiva e geoespacial requer dados confi√°veis. Outliers podem distorcer modelos estat√≠sticos e previs√µes, afetando o crit√©rio de sucesso (85% de precis√£o nos rios de maior risco). O IQR foi escolhido por ser robusto a distribui√ß√µes assim√©tricas, comuns em dados ambientais e socioecon√¥micos.
- **Valida√ß√£o de Dados**: Dados de polui√ß√£o pl√°stica e risco ambiental podem conter erros de medi√ß√£o ou entradas excepcionais (ex.: rios industriais). Identificar outliers permite validar se s√£o reais (ex.: impacto de grandes popula√ß√µes) ou erros a serem corrigidos.
- **Prepara√ß√£o para Modelagem**: Antes de criar modelos preditivos (ex.: para prever `Plastic_to_River_2060_tons` ou agrupar pa√≠ses por risco), √© essencial tratar outliers para evitar vi√©s nos resultados, especialmente nas vari√°veis derivadas como `Plastic_Growth_Rate` (120 outliers) e `Relative_Risk` (391 outliers).
- **Relev√¢ncia para Perguntas de Neg√≥cio**: Outliers em `Plastic_Growth_tons` e `Plastic_to_River_2060_tons` s√£o diretamente ligados √†s perguntas 1 (crescimento de pl√°sticos) e 3 (rios de maior risco), exigindo an√°lise para garantir precis√£o nas recomenda√ß√µes.

### Por que IQR e n√£o Outros M√©todos?
- Comparado ao desvio padr√£o (que assume normalidade), o IQR √© mais adequado para datasets com distribui√ß√£o n√£o normal, como os de polui√ß√£o e popula√ß√£o.
- Diferente de m√©todos baseados em percentis fixos, o IQR adapta-se √† variabilidade interna dos dados, oferecendo uma abordagem flex√≠vel para conjuntos diversificados como este.

## Conclus√£o
O uso do m√©todo IQR foi necess√°rio para garantir a qualidade dos dados antes de avan√ßar para a modelagem preditiva e a gera√ß√£o de insights geoespaciais. A identifica√ß√£o de outliers em vari√°veis cr√≠ticas sugere a necessidade de uma an√°lise detalhada por `Country` e `River_Name` para decidir se esses valores devem ser mantidos, removidos ou capeados nos limites IQR, alinhando-se aos objetivos do projeto.

# Documenta√ß√£o das T√©cnicas Estat√≠sticas

## 1. An√°lise de Crescimento Percentual
**T√©cnica Utilizada**: C√°lculo de varia√ß√£o percentual entre 2015-2060  
**Por que foi usada**: Para identificar √°reas com maior aumento projetado de polui√ß√£o pl√°stica  
**Como interpretar**:
- Valores positivos indicam aumento na emiss√£o de pl√°sticos
- Valores negativos indicam redu√ß√£o
- Quanto maior o percentual, mais cr√≠tica a situa√ß√£o
- Pa√≠ses/rios no topo da lista s√£o prioridades para interven√ß√£o

**F√≥rmula**:  
`Crescimento Percentual = ((Valor2060 - Valor2015) / Valor2015) √ó 100`

---

## 2. An√°lise de Correla√ß√£o
**T√©cnica Utilizada**: Correla√ß√£o de Spearman + Regress√£o Linear  
**Por que foi usada**: Para entender rela√ß√µes entre vari√°veis socioecon√¥micas e polui√ß√£o  
**Como interpretar**:
- Coeficientes entre -1 e 1
- |r| > 0.7: forte correla√ß√£o
- 0.5 < |r| ‚â§ 0.7: moderada
- |r| ‚â§ 0.5: fraca
- Sinais indicam dire√ß√£o da rela√ß√£o (positiva/negativa)

**Visualiza√ß√£o**: Matriz de correla√ß√£o com heatmap

---

## 3. Identifica√ß√£o de Rios Resilientes
**T√©cnica Utilizada**: Filtragem por condi√ß√µes m√∫ltiplas + An√°lise descritiva  
**Por que foi usada**: Para encontrar rios que mant√™m alto risco apesar de pol√≠ticas de mitiga√ß√£o  
**Como interpretar**:
- Foco em rios com:
  - `Plastic_Risk_Category = 'Alto'`
  - `Policy_Strength_2060 > 0.7`
- Valores altos de `Plastic_to_River_2060_tons` indicam resili√™ncia ao risco
- Requerem estrat√©gias espec√≠ficas al√©m das convencionais

---

## 4. Modelo Preditivo
**T√©cnica Utilizada**: Regress√£o Linear M√∫ltipla  
**Por que foi usada**: Para prever carga pl√°stica com base em indicadores socioecon√¥micos  
**Como interpretar**:
- Coeficientes positivos: vari√°vel aumenta a carga pl√°stica
- Coeficientes negativos: vari√°vel reduz a carga pl√°stica
- Magnitude indica for√ßa do efeit
- R¬≤ mostra poder explicativo do modelo (0-1)

**Vari√°veis inclu√≠das**:
1. PIB per capita
2. Popula√ß√£o
3. Taxa de coleta de lixo
4. For√ßa de pol√≠ticas ambientais

---

## 5. Segmenta√ß√£o de Pa√≠ses
**T√©cnica Utilizada**: Clusteriza√ß√£o K-Means  
**Por que foi usada**: Para agrupar pa√≠ses com perfis similares de risco/potencial de a√ß√£o  
**Como interpretar**:

  1. Alto risco/baixa a√ß√£o

  3. Baixo risco/alta a√ß√£o

**Vari√°veis de clusteriza√ß√£o**:
